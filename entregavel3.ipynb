**LETRA A**

Retirar a coluna "employee_id" ja que nao faz diferenca para a analise

```python
df = pd.read_csv("dataset3.psv", sep="|")
df = df.drop(columns=[" employee_id"])
```

```python
df
```

Corrigindo as entradas da coluna de genero

```python
df['gender'] = df['gender'].replace({'MALE': 'm', 'fEMALE': 'f', 'eu sou homem hetero homossexual!': 'm'})
df
```

Corrigindo os valores nao numericos

```python
coluna1 = 'no_of_trainings'
media1 = df[coluna1].mean()
df[coluna1].fillna(media1, inplace=True)

coluna2 = 'previous_year_rating'
media2 = df[coluna2].mean()
df[coluna2].fillna(media2, inplace=True)

coluna3 = 'avg_training_score'
media3 = df[coluna3].mean()
df[coluna3].fillna(media3, inplace=True)

df
```

Corrigindo os valores extremamente altos de idade

```python
media_idades = df.loc[df["age"] <= 120000, "age"].mean()
df.loc[df["age"] > 120000, "age"] = media_idades
df
```

**LETRA E**

Categorizando a coluna de gÃªnero, departamento e regiÃ£o para ser possivel treinar o modelo

```python
df = pd.get_dummies(df, columns=['department'])
df = pd.get_dummies(df, columns=['region'])
df = pd.get_dummies(df, columns=['gender'])
df
```

Normalizando as colunas de anos de trabalho e avaliacao no ano anterior

```python
min_ano = df['length_of_service'].min()
max_ano = df['length_of_service'].max()
df['length_of_service'] = (df['length_of_service'] - min_ano) / (max_ano - min_ano)

min_ant = df['previous_year_rating'].min()
max_ant = df['previous_year_rating'].max()
df['previous_year_rating'] = (df['previous_year_rating'] - min_ant) / (max_ant - min_ant)

df
```

```python
df = df.drop(columns=["age"])
df = df.drop(columns=["avg_training_score"])

df
```

```python
min_num = df['no_of_trainings'].min()
max_num = df['no_of_trainings'].max()
df['no_of_trainings'] = (df['no_of_trainings'] - min_num) / (max_num - min_num)

df
```

Agora todos os dados estÃ£o prontos para treinar o modelo

```python
from sklearn.model_selection import train_test_split

X = df.iloc[:, 1:]
y = df['is_promoted']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("AcurÃ¡cia:", accuracy)
```

O menor nÃºmero de vizinhos para tal acurÃ¡cia Ã© dois.

**LETRA F**

Um possÃ­vel erro durante o treinamento Ã© que a acurÃ¡cia estÃ¡ muito alta, o que nÃ£o necessariamente Ã© algo positivo. Nessa base de dados, existem muito mais dados para pessoas que nÃ£o foram promovidas em comparaÃ§Ã£o com as que foram promovidas, causando um problema de dados desbalanceados, o que acaba inviesando a base de dados e corrompendo o processo de aprendizado de mÃ¡quina.

**LETRA G**

```python
from sklearn.metrics import recall_score

precisao = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("PrecisÃ£o: ", precisao)
print("Recall:", recall)
```

Esses valores indicam que o modelo tem maior capacidade de classificar exemplos negativos corretamente do que exemplos positivos corretamente, o que condiz com o excesso de dados negativos que levou aos desbalanceamento de dados citado anteriormente.
